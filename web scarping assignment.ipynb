```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the ICC ODI Men's Team Rankings page
url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Find the table containing the ODI team rankings
    table = soup.find("table", class_="table")
    
    # Initialize empty lists to store data
    teams = []
    matches = []
    points = []
    ratings = []

    # Find all rows in the table
    rows = table.find_all("tr")[1:11]  # Extract top 10 teams

    # Iterate through the rows and extract the required data
    for row in rows:
        cols = row.find_all("td")
        team_name = cols[1].text.strip()
        match_count = cols[2].text
        point_count = cols[3].text
        rating = cols[4].text.strip()

        # Append the data to the respective lists
        teams.append(team_name)
        matches.append(match_count)
        points.append(point_count)
        ratings.append(rating)

    # Create a DataFrame
    data = {
        "Team": teams,
        "Matches": matches,
        "Points": points,
        "Rating": ratings
    }
    df = pd.DataFrame(data)

    # Print the DataFrame
    print(df)

else:
    print("Failed to retrieve data from the ICC website.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the ICC ODI Batsmen Rankings
url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table containing the rankings data
table = soup.find('table', class_='table rankings-table')

# Initialize empty lists to store data
players = []
teams = []
ratings = []

# Loop through the rows of the table to extract data
for row in table.find_all('tr')[1:11]:  # Skip the header row and limit to top 10
    cols = row.find_all('td')
    players.append(cols[1].get_text(strip=True))
    teams.append(cols[2].get_text(strip=True))
    ratings.append(cols[3].get_text(strip=True))

# Create a DataFrame
data = {'Player': players, 'Team': teams, 'Rating': ratings}
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the ICC Women's ODI rankings page
url = "https://www.icc-cricket.com/rankings/womens/team-rankings/odi"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the table containing the rankings data
    table = soup.find('table', {'class': 'table'})

    # Initialize empty lists to store data
    teams = []
    matches = []
    points = []
    ratings = []

    # Find all the rows in the table (excluding the header row)
    rows = table.find_all('tr')[1:]

    # Loop through the rows and extract data
    for row in rows:
        columns = row.find_all('td')
        if len(columns) >= 5:
            team = columns[1].text.strip()
            match = columns[2].text.strip()
            point = columns[3].text.strip()
            rating = columns[4].text.strip()
            teams.append(team)
            matches.append(match)
            points.append(point)
            ratings.append(rating)

    # Create a DataFrame from the extracted data
    data = {
        'Team': teams,
        'Matches': matches,
        'Points': points,
        'Rating': ratings
    }
    df = pd.DataFrame(data)

    # Display the top 10 teams
    top_10_teams = df.head(10)
    print(top_10_teams)

else:
    print("Failed to retrieve the data. Status code:", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL of the webpage to scrape
url = "https://currentaffairs.adda247.com/list-of-presidents-of-india/"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")

    # Find the table containing the list of presidents
    president_table = soup.find("table")

    # Initialize empty lists to store the data
    names = []
    terms = []

    # Loop through the rows of the table
    for row in president_table.find_all("tr")[1:]:  # Skip the header row
        columns = row.find_all("td")
        if len(columns) >= 2:
            name = columns[0].text.strip()
            term = columns[1].text.strip()
            names.append(name)
            terms.append(term)

    # Create a DataFrame from the scraped data
    df = pd.DataFrame({
        "Name": names,
        "Term of Office": terms
    })

    # Print the DataFrame
    print(df)

else:
    print("Failed to retrieve the webpage")

```
